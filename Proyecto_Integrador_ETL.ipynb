### Verificamos registros duplicados


```python
filas_duplicadas = datos_pacientes[datos_pacientes.duplicated()]
print("Filas duplicadas:")
print(filas_duplicadas)
```

```python
datos_pacientes_sin_duplicados = datos_pacientes.drop_duplicates().reset_index(drop=True)
datos_pacientes_sin_duplicados
```

```python
datos_pacientes_sin_duplicados.shape
```

Observamos que despues de eliminar los registros duplicados obtenemos un total de 461 registros

### Verificamos datos faltantes

```python
print(datos_pacientes_sin_duplicados.isnull().sum())
```

```python
porcentaje_nulos_por_columna = round((datos_pacientes_sin_duplicados.isnull().sum() / datos_pacientes_sin_duplicados.shape[0]) * 100,2)
print("Porcentaje de valores nulos por columna:")
print(porcentaje_nulos_por_columna)
```

```python
cantidad_registros_nulos = len(datos_pacientes_sin_duplicados[datos_pacientes_sin_duplicados.isna().any(axis=1)])
porcentaje = round((cantidad_registros_nulos / len(datos_pacientes_sin_duplicados)) * 100, 2)
print(f'Hay {cantidad_registros_nulos} registros con al menos un valor nulo, lo que representa el {porcentaje}% del total de registros.')
```

```python
nuevo_nombre = "datos_modificados"
datos_modificados = datos_pacientes_sin_duplicados.copy() 
datos_modificados.name = nuevo_nombre
```

```python
datos_modificados =datos_modificados.dropna()
datos_modificados
```

```python
print(datos_modificados.isnull().sum())
```

```python
datos_modificados.reset_index(inplace=True, drop=True)
datos_modificados
```

```python
datos_modificados["EDAD"]= datos_modificados["EDAD"].astype(int)
datos_modificados["DIAS HOSPITALIZACION MQ"]= datos_modificados["DIAS HOSPITALIZACION MQ"].astype(int)
datos_modificados["NUMERO DE MUESTRAS TOMADAS"]= datos_modificados["NUMERO DE MUESTRAS TOMADAS"].astype(int)
datos_modificados["DIAS HOSPITALIZACIÓN UPC"]= datos_modificados["DIAS HOSPITALIZACIÓN UPC"].astype(int)
```

```python
datos_modificados["NUMERO DE DIAS POST BIOPSIA EN QUE SE PRESENTA LA COMPLICACIÓN INFECCIOSA"] = datos_modificados["NUMERO DE DIAS POST BIOPSIA EN QUE SE PRESENTA LA COMPLICACIÓN INFECCIOSA"].replace("NO", 0)
```

```python
datos_modificados["ANTIBIOTICO UTILIAZADO EN LA PROFILAXIS"]= datos_modificados["ANTIBIOTICO UTILIAZADO EN LA PROFILAXIS"].replace('FLUOROQUINOLONA_AMINOGLICÓSIDO','FLUOROQUINOLONA_AMINOGLICOSIDO')
```

```python
datos_modificados["ENF. CRONICA PULMONAR OBSTRUCTIVA"]= datos_modificados["ENF. CRONICA PULMONAR OBSTRUCTIVA"].replace('SI, ASMA','SI')
datos_modificados["ENF. CRONICA PULMONAR OBSTRUCTIVA"]= datos_modificados["ENF. CRONICA PULMONAR OBSTRUCTIVA"].replace('SI, EPOC','SI')
```

```python
datos_modificados["BIOPSIA"].unique()
```

```python
datos_modificados["BIOPSIA"] = datos_modificados["BIOPSIA"].str.rstrip()
```

```python
datos_modificados["PATRON DE RESISTENCIA"].unique()
```

```python
datos_modificados["PATRON DE RESISTENCIA"]= datos_modificados["PATRON DE RESISTENCIA"].replace(' AMPI R, CIPRO R, GENTA R, SULFA M R','SI')
datos_modificados["PATRON DE RESISTENCIA"]= datos_modificados["PATRON DE RESISTENCIA"].replace('RESISTENTE A AMPI, CIPRO Y GENTA','SI')
datos_modificados["PATRON DE RESISTENCIA"]= datos_modificados["PATRON DE RESISTENCIA"].replace('RESISTENTE A AMPI, SULFA, CEFADROXILO, CEFUROXIMO, CIPRO Y CEFEPIME, CEFOTAXIMA','SI')
datos_modificados["PATRON DE RESISTENCIA"]= datos_modificados["PATRON DE RESISTENCIA"].replace('MULTI SENSIBLE','SI')
```

```python
datos_modificados = datos_modificados.rename(columns={'EDAD': 'Edad','DIABETES': 'Diabetes','HOSPITALIZACIÓN ULTIMO MES': 'Hosp_ultimo_mes', 'PSA': 'PSA',
                                                  'BIOPSIAS PREVIAS': 'Biopsias_previas','VOLUMEN PROSTATICO': 'Vol_prostatico',
                                                  'ANTIBIOTICO UTILIAZADO EN LA PROFILAXIS': 'Antibiotico_profilaxis',
                                                  'NUMERO DE MUESTRAS TOMADAS': 'Num_muestras','CUP': 'CUP',
                                                  'ENF. CRONICA PULMONAR OBSTRUCTIVA': 'Enf_cronica_pulm_obstructiva',
                                                  'BIOPSIA': 'Biopsia',
                                                  'NUMERO DE DIAS POST BIOPSIA EN QUE SE PRESENTA LA COMPLICACIÓN INFECCIOSA': 'Dias_post_biopsia_infeccion',
                                                   'FIEBRE': 'Fiebre','ITU': 'ITU',
                                                   'TIPO DE CULTIVO': 'Tipo_cultivo',
                                                   'AGENTE AISLADO': 'Agente_aislado',
                                                   'PATRON DE RESISTENCIA': 'Patron_resistencia','HOSPITALIZACION': 'Hospitalizacion',
                                                   'DIAS HOSPITALIZACION MQ': 'Dias_hosp_MQ',
                                                   'DIAS HOSPITALIZACIÓN UPC': 'Dias_hosp_UPC'
                                                  })
```
