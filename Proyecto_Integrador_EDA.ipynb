## 6. Codificacion del dataset

```python
df_codificado = df.copy()
```

```python
mapeo = {"NO": 0, "SI": 1}
columnas_a_mapear = ["Diabetes","Biopsias_previas","Vol_prostatico","Enf_cronica_pulm_obstructiva","Fiebre","ITU","Patron_resistencia","Hospitalizacion"]
df_codificado[columnas_a_mapear] = df_codificado[columnas_a_mapear].apply(lambda x: x.map(mapeo))
```

```python
orden = {"FLUOROQUINOLONA_AMINOGLICOSIDO": 0, "CEFALOSPORINA_AMINOGLUCOCIDO": 1,"OROQUINOLONAS": 2,"OTROS": 3}
df_codificado["Antibiotico_profilaxis"] = df_codificado["Antibiotico_profilaxis"].map(orden)
```

```python
orden1 = {"NO": 0, "HEMOCULTIVO": 1,"UROCULTIVO": 2,"HEMOCULTIVO Y UROCULTIVO": 3}
df_codificado["Tipo_cultivo"] = df_codificado["Tipo_cultivo"].map(orden1)
```

```python
orden2 = {"NO": 0, "E.COLI": 1,"PSEUDOMONAS AERUGINOSA": 2}
df_codificado["Agente_aislado"] = df_codificado["Agente_aislado"].map(orden2)
```

```python
orden3 = {"NEG": 0, "ADENOCARCINOMA GLEASON 6": 1,"ADENOCARCINOMA GLEASON 7": 2,"ADENOCARCINOMA GLEASON 10": 3,"ADENOCARCINOMA GLEASON 9": 4,
          "ADENOCARCINOMA GLEASON 8": 5, "PROSTATITIS": 6, "HIPERPLASIA PROSTATICA": 7, "CARCINOMA INDIFERENCIADO DE CELULAS CLARAS": 8}
df_codificado["Biopsia"] = df_codificado["Biopsia"].map(orden3)
```

## Matriz de Correlación

```python
matriz_correlacion = df_codificado.corr()
plt.figure(figsize=(10,10))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Matriz de Correlación")
plt.show()
```

Si hay una fuerte correlación entre estas variables, la eliminación de algunas de ellas puede ayudar a reducir la redundancia en tu modelo y mejorar la interpretabilidad
* Correlación entre 'Fiebre', 'Dias_post_biopsia_infeccion' y 'Hospitalizacion': se observa una alta correlacion entre las variables por lo que se elimina la variable 'Dias_post_biopsia_infeccion' que indica los dias posteriores a la biopsia en la que se presenta infeccion, dado que es redundante teniendo las otras variables

* Relación entre 'Agente_aislado', 'Tipo_cultivo', 'Patron_resistencia' y'ITU' : en este caso la correlacion es muy elevada por lo que para no tener redundancia eliminaremos las variables 'Agente_aislado', 'Patron_resistencia' y 'Tipo_cultivo'

```python
df = df.drop(columns=["Agente_aislado"])
df = df.drop(columns=["Patron_resistencia"])
df = df.drop(columns=["Dias_post_biopsia_infeccion"])
df = df.drop(columns=["Tipo_cultivo"])

df_codificado = df_codificado.drop(columns=["Agente_aislado"])
df_codificado = df_codificado.drop(columns=["Patron_resistencia"])
df_codificado = df_codificado.drop(columns=["Dias_post_biopsia_infeccion"])
df_codificado = df_codificado.drop(columns=["Tipo_cultivo"])
```

```python
matriz_correlacion = df_codificado.corr()
plt.figure(figsize=(10,10))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Matriz de Correlación 2")
plt.show()
```
